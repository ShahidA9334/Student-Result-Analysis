```python
import pandas as pd
import numpy as np
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt

# Upload the file to the current session
uploaded = files.upload()

# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]  # Assuming only one file is uploaded

# Read the CSV file using the uploaded filename
dataset = pd.read_csv(filename)
```

```python
dataset = dataset.drop("Unnamed: 0", axis = 1)
dataset.head()
```

Droped Unnamded column

```python
dataset["WklyStudyHours"] = dataset["WklyStudyHours"].str.replace("05-Oct", "5-10")
dataset.head(5)
```

change weekly study hours column

```python
plt.figure(figsize = (5, 5))
ax = sns.countplot(x = "Gender", data = dataset)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```


gender distribution

```python
gb = dataset.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
gb
```

from the above chart we ahve analysed that: the number of females in the data is more than the number of males

```python
plt.figure(figsize = (5, 5))
sns.heatmap(gb, annot=True)
plt.title("Relationship between parent's and student's score")
plt.show()
```

from the above chart we have concluded that use eduction of the parents have a good impact in this course

```python
gb1 = dataset.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
gb1
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between parent's Marital Status and student's score")
plt.show()
```

from the above chart we have concluded that there is no/negligible impact on the student's score due tp their parent's marital status

Distribution of Ethnic Groups

```python
groupA = dataset.loc[dataset['EthnicGroup'] == 'group A'].count()
groupB = dataset.loc[dataset['EthnicGroup'] == 'group B'].count()
groupC = dataset.loc[dataset['EthnicGroup'] == 'group C'].count()
groupD = dataset.loc[dataset['EthnicGroup'] == 'group D'].count()
groupE = dataset.loc[dataset['EthnicGroup'] == 'group E'].count()

l = ['group A', 'group B', 'group C', 'group D', 'group E']
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mylist)
plt.figure(figsize = (5, 5))
plt.pie(mylist, labels = l, autopct = '%1.1f%%')
plt.title("Ethnic Group Distribution")
plt.show()
```
